# Description des colonnes :
## des information sur le sol : 

#### type_sol
    Type de sol (argileux, sableux...)
Sert à déterminer la capacité du sol à retenir l’eau
#### ph_sol
     Niveau d’acidité du sol
Influence l’absorption des nutriments et de l’eau
#### humidite_sol
     Quantité d’eau présente dans le sol
Indicateur principal pour décider l’irrigation
#### carbone_organique
    Matière organique dans le sol
Améliore la fertilité et la rétention d’eau
#### conductivite_electrique
    Niveau de salinité du sol
Un sol trop salé peut nuire à la plante

## Conditions climatiques
#### temperature
    Température ambiante
Influence l’évaporation de l’eau
#### humidite
    Humidité de l’air
Affecte la perte d’eau par évaporation
#### precipitation
    Quantité de pluie
Peut réduire ou remplacer l’irrigation
#### heures_ensoleillement
    Nombre d’heures de soleil
Plus de soleil = plus d’évaporation
#### vitesse_vent
    Vitesse du vent
Accélère le séchage du sol
## Culture
#### type_culture
    Type de plante cultivée
Chaque culture a des besoins en eau différents
#### stade_croissance
    Stade de développement de la plante
Les besoins en eau varient selon le stade
#### saison
    Saison de culture
Impacte les conditions climatiques
## Irrigation
### type_irrigation
    Méthode d’irrigation (goutte-à-goutte, aspersion...)
Influence l’efficacité de l’eau utilisée
#### source_eau
    Origine de l’eau (puits, rivière...)
Peut influencer la disponibilité
#### surface_champ
    Taille du champ (en hectares)
Détermine la كمية totale d’eau nécessaire
#### paillage
    Utilisation de paillage (oui/non)
Réduit l’évaporation et conserve l’humidité
#### irrigation_precedente
    Quantité d’eau utilisée précédemment
Aide à éviter le sur-arrosage

```python
df = df.drop_duplicates()
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
# supprition des valeurs aberantes 
df = df[df['temperature'] < 60]
df = df[(df['humidite_sol'] >= 0) & (df['humidite_sol'] <= 100)]
df.info()
```

```python
# normalisation et la standardisation 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['temperature','humidite_sol','humidite']] = scaler.fit_transform(
    df[['temperature','humidite_sol','humidite']]
)
```

```python
#pour transformer les variable categorielles on des variables numeriques
df = pd.get_dummies(df, drop_first=True) 
```

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
# Étape 1 : Créer une seule colonne "target" à partir des colonnes encodées
# idxmax(axis=1) permet de récupérer le nom de la colonne qui contient la valeur 1 (la classe active)
df['target'] = df[['besoin_irrigation_Low', 'besoin_irrigation_Medium']].idxmax(axis=1)


# Étape 2 : Transformer les valeurs textuelles en valeurs numériques
# On convertit :
# 'besoin_irrigation_Low' → 0
# 'besoin_irrigation_Medium' → 1
df['target'] = df['target'].map({
    'besoin_irrigation_Low': 0,
    'besoin_irrigation_Medium': 1
})


# Étape 3 : Supprimer les anciennes colonnes (one-hot encoding)
# Ces colonnes ne sont plus nécessaires après la création de la target
df = df.drop(['besoin_irrigation_Low', 'besoin_irrigation_Medium'], axis=1)


# Étape 4 (optionnelle) : Vérification
# Afficher les premières lignes pour confirmer que tout est correct
df.head()
```

```python
# transformer les colonnes que contien la variable bool en int 
bool_cols = df.select_dtypes(include=['bool']).columns

df[bool_cols] = df[bool_cols].astype(int)
```
